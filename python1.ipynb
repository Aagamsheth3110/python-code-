```python
### pie chart ###
import matplotlib.pyplot as plt
import numpy as np
company_name = ["infosys","ADANI","AMBANI","TATA"]
production=np.array([20,40,65,80])
colour=['b','w','y','p']
plt.pie(production,labels= company_name,explode=(0,0.1,0,0.1),autopct='%1.1f%%',shadow=True,startangle=None,colors=None,radius=0.9,frame=False,rotatelabels=True)
plt.legend()
plt.show()
```
